# Linear Regression Exercise

This notebook performs a simple linear regression on a given dataset of X and Y values.

**Tasks:**
1. Create a function to predict the y-value given any x-value (`def y_pred()`)
2. Create a function to calculate the error between predicted and actual values using MSE los (`def loss_function()`)
3. Plot a scatter diagram of the data-points and a line chart of the predicted values.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Given data:

X = [0., 0.52631579, 1.05263158, 1.57894737, 2.10526316, 2.63157895, 3.15789474, 3.68421053, 4.21052632, 4.73684211, 5.26315789,
     5.78947368, 6.31578947, 6.84210526, 7.36842105, 7.89473684, 8.42105263, 8.94736842, 9.47368421, 10.]

Y = [2.60943416, 1.76213705, 7.18511292, 9.40744522, 5.46635068, 8.6061673, 13.30831239, 14.26225166, 16.70323979, 16.87285951,
    22.17984858, 23.81874177, 24.23732455, 28.20185083, 28.72449237, 27.91299402, 32.21118578, 31.39802427, 36.91479534, 36.90014818]

# Converting to numpy arrays:
x = np.array(X)
y = np.array(Y)
```

### Calculating the slope (m) and constant (c)
We use the formulas:

$
m = \frac{n\sum(xy) - \sum x \sum y}{n\sum(x^2) - (\sum x)^2}
$

$
c = \frac{\sum y - m \sum x}{n}
$

```python
n = len(x)
m = (n * np.sum(x*y) - np.sum(x)*np.sum(y)) / (n* np.sum(x**2) - (np.sum(x))**2)
c = (np.sum(y) - m*np.sum(x)) / n
print("Slope (m):", m)
print("Constant (c):", c)
```

### Define functions for prediction and error calculation

```python
# Answer 1: Function to predict the y value given any x value
def y_pred(x, m=m, c=c):
    return m*x + c

#Answer 2: Function to calculate the errors between actual and predicted values.
def loss_function(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
# Predict and Calculating error:
y_hat = y_pred(x)
error = loss_function(y, y_hat)
print("Mean Squared Error:", error)
```

### Scatter Diagram of Data Points and Regression Line

```python
# Answer 3: Plotting a scatter diagram of the data points and a line chart of the predicted values
plt.scatter(x,y, label = "Actual Data")
plt.plot(x, y_hat, color='red', label='Predicted Line')
plt.title("Linear Regression Fit")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
